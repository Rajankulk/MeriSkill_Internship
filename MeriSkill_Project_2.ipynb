```python
#Removing Outliers
Average = diabetes["BloodPressure"].mean()
diabetes["BloodPressure"] = diabetes["BloodPressure"].apply(lambda v:Average if v==0 else v)
```

```python
plt.figure(figsize = (8,5))
sns.distplot(diabetes["BloodPressure"] , bins = 30,color = "black")
plt.title("Blood-Pressure Distribution")
```

```python
##Z score to check outliers in  features.
diabetes["z_score_BP"]=(diabetes.BloodPressure-diabetes.BloodPressure.mean()/diabetes.BloodPressure.std())
diabetes["z_score_Glucose"]=(diabetes.Glucose-diabetes.Glucose.mean()/diabetes.Glucose.std())
```

```python
diabetes.head()
```

```python
diabetes.describe()
```

```python
#dropping unnecessary columns
diabetes.drop(columns=["z_score_BP", "z_score_Glucose"],inplace=True)
```

```python
diabetes.head()
```

```python
#split data into train and test
X = diabetes.drop("Outcome", axis=1)
Y = diabetes['Outcome']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=30)
```

```python
x_train.shape , x_test.shape 
```

```python
y_train.shape , y_test.shape
```

```python
#Training model
#LR
Train_Model = LogisticRegression()
Train_Model.fit(x_train,y_train)
```

```python
#Making predictions #y predicted
predict = Train_Model.predict(x_test)
```

```python
print(predict)
```

```python
Accuracy = accuracy_score(predict,y_test)
```

```python
print("Accuracy:", Accuracy)
```

```python
#Model Building & Evaluating model performance
from sklearn.metrics import confusion_matrix
Confusion_matrix = confusion_matrix(y_test,predict)
print('Confusion_matrix: \n',Confusion_matrix)
```

```python
#classification report
from sklearn.metrics import classification_report
Classification_report = classification_report(y_test,predict)
print('Classification_report: \n',Classification_report)
```

```python
from sklearn.metrics import recall_score,precision_score,f1_score

Recall = recall_score(y_test, predict)
print('Recall:',Recall)

Precision = precision_score(y_test, predict)
print('Precision:',Precision)

F1_score = f1_score(y_test,predict)
print('F1_score:',F1_score)
```
